```python
import pandas as pd

#df = pd.read_csv(OPSD_PATH, parse_dates=["utc_timestamp"], index_col="utc_timestamp")
df = pd.read_csv("../data/time_series_60min_singleindex.csv", parse_dates=["utc_timestamp"], index_col="utc_timestamp")
print(df.shape)
#df.head(20)
df.tail(20)
```

```python
# Check the columns with DE in their name
df_contains_DE = df.filter(like="DE_")
print(df_contains_DE.columns.tolist())

# Include the DE cloumns to sudy for the Germany dataset
core_cols = [ "DE_load_actual_entsoe_transparency", "DE_load_forecast_entsoe_transparency", "DE_solar_generation_actual","DE_wind_generation_actual",]

# Print the trailing last DE columns
df_de = df[core_cols].copy()
#df_de.head(20)
df_de.tail(20)
df_de.plot(subplots=True, figsize=(12,10))

#print the total number of nans present for each column
df_de.isna().sum()
```

```python
import matplotlib.pyplot as plt

# Ensure index is datetime
df_de = df_de.sort_index()
if not isinstance(df_de.index, pd.DatetimeIndex):
    df_de.index = pd.to_datetime(df_de.index)

# NaN before check
print("NaNs before fill \n", df_de.isna().sum())

# fill small gaps by time interpolation
# df_de_filled = df_de.interpolate(method="time")
df_de_filled = df_de.interpolate(method="time", limit_direction="both")

# Safety net (in case NaNs remain at edges)
df_de_filled = df_de_filled.ffill().bfill()

# NaN after check
print("\nNaNs after filling procedure \n", df_de_filled.isna().sum())

# check correlations
corr = df_de_filled.corr()
print("\nCorrelation matrix:\n", corr.round(3))

import seaborn as sns
plt.figure(figsize=(9,7))
sns.heatmap(corr, annot=True, fmt=".1f", cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Heatmap (This is filled series)")
plt.show()
```

```python
# Further look through data, residuals, operational changes etc
residual_1 = df_de["DE_load_actual_entsoe_transparency"]-df_de["DE_load_forecast_entsoe_transparency"]
residual_1.plot(figsize=(12,4))
```

```python
regular_mean = df_de["DE_load_actual_entsoe_transparency"].groupby(df_de.index.hour).mean()
residual_2 = df_de["DE_load_actual_entsoe_transparency"] - regular_mean[df_de.index.hour].values
residual_2.plot(figsize=(12,4))
```

```python
multivar_rat_1= df_de["DE_solar_generation_actual"]/ df_de["DE_load_actual_entsoe_transparency"]
multivar_rat_1.plot(figsize=(12,4))
```

```python
multivar_rat_2 = df_de["DE_wind_generation_actual"]/ df_de["DE_load_actual_entsoe_transparency"]
multivar_rat_2.plot(figsize=(12,4))
```

```python
# basic z-score analysis:

residual = df_de["DE_load_actual_entsoe_transparency"] - df_de["DE_load_forecast_entsoe_transparency"]
window = 24*7  # 1 week baseline
res_mean = residual.rolling(window).mean()
res_std  = residual.rolling(window).std()
residual_z = (residual-res_mean)/res_std

#pd.Series(residual_z).hist(bins=100, density=True)
plt.figure(figsize=(12, 4))
plt.plot(residual_z, label="residual z-score", alpha=0.6)
plt.axhline(3,  color="black", linestyle="-", label="+3 sigma")
plt.axhline(-3, color="black", linestyle="-", label="-3 sigma")
plt.title("Standardized load residuals (rolling z-score)")
plt.ylabel("z-score")
plt.legend()
plt.show()

#Highlight anamolies explicitly
anomalies = residual_z.abs() > 3
plt.figure(figsize=(12, 4))
plt.plot(residual_z, label="residual z-score", alpha=0.6)
#plt.scatter(residual_z.index[anomalies], residual_z[anomalies], color="red", label="Anomalies")
plt.scatter(residual_z.index[anomalies], residual_z[anomalies], color="red", s=6, label="Anomalies")
plt.axhline(3, color="black", linestyle="-")
plt.axhline(-3, color="black", linestyle="-")
plt.title("Detected potential anomalies via z-score")
plt.legend()
plt.show()
```
