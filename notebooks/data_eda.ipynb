```python
# Basline ML model (iso forests)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Sort the dataset in terms of date-time approach 
df_de = df_de.sort_index()
# Redifining this again just for having a cell-flow
residual=(df_de["DE_load_actual_entsoe_transparency"]-df_de["DE_load_forecast_entsoe_transparency"]).astype(float)
# filling + sanity as done above for NaNs
residual = residual.interpolate(method="time").ffill().bfill()

#can clip insane values (seems rare, commenting this as of now)
# residual = residual.clip(lower=residual.quantile(0.001), upper=residual.quantile(0.999))

###############################################################
# create residual-only feature matrix, but with added temporal context
##############################################################
X=pd.DataFrame(index=residual.index)
X["resid"] = residual

#add lag features: 1h, 6h, 24h lags (since OPSD is hourly)
for lag in [1, 6, 24]:
    X[f"resid_lag_{lag}"]=residual.shift(lag)
# Rolling stats
X["resid_roll_mean_24"]=residual.rolling(24).mean()
X["resid_roll_std_24"]= residual.rolling(24).std()

# Drop initial NaNs from lag/rolling
X = X.dropna()

###############################################################
# train isolation forest, contamination is the expected anomaly fraction
##############################################################
contamination=0.01  # 1% anomalies as a starting point
iso=IsolationForest(n_estimators=400, contamination=contamination, random_state=42)
iso.fit(X)
#TO-DO: Split and test, 
#############################################
#convert model output into anamolous signals
#############################################
# decision_function:higher is more normal, lower is more anomalous
scores = pd.Series(iso.decision_function(X), index=X.index, name="iforest_score")
# predict: -1 anomaly, +1 normal
pred = pd.Series(iso.predict(X), index=X.index, name="iforest_pred")
is_anom = (pred == -1).astype(int).rename("is_anomaly_iforest")

#############################################################
# convert point anomalies to  event anomalies (clusters), 
# merge anomalies that are close together (gap tolerance)
############################################################
events_df=pd.concat([residual.reindex(X.index).rename("residual"), scores, is_anom], axis=1)

# Parameters to merge nearby points into the same event
gap_tolerance_hours=2  # if anomalies separated by <=2 hours, treat as same event

# Identify start of new anomaly group:
# new group if (current is anomaly) and (previous anomaly is too far)
anom_times=events_df.index[events_df["is_anomaly_iforest"] == 1]

if len(anom_times)==0:
    print("No anomalies detected. Try with increasing contamination!")
else:
    # Build group ids based on time gaps
    time_diffs=anom_times.to_series().diff().dt.total_seconds().div(3600)
    new_group = (time_diffs.isna()) | (time_diffs > gap_tolerance_hours)
    group_id = new_group.cumsum().values
    # Aggregate event metrics
    anom_only= events_df.loc[anom_times].copy()
    anom_only["event_id"]=group_id
    # Event severity metrics (residual in MWatts)
    event_summary = (anom_only.groupby("event_id").agg(
            start_time=("residual", lambda s: s.index.min()),
            end_time=("residual", lambda s: s.index.max()),
            duration_hours=("residual", "count"),
            max_abs_residual_MW=("residual", lambda s: float(np.max(np.abs(s.values)))),
            mean_abs_residual_MW=("residual", lambda s: float(np.mean(np.abs(s.values)))),
            min_score=("iforest_score", "min"))
        .sort_values("start_time")
        .reset_index(drop=True))

    ######################################################
    #frequency, duration, severity over time
    #####################################################
    event_summary["month"]=event_summary["start_time"].dt.to_period("M").dt.to_timestamp()
    # frequency: events per month
    events_per_month=event_summary.groupby("month").size().rename("events_per_month")
    # Duration: distribution + mean per month
    duration_stats = event_summary["duration_hours"].describe()
    # Severity trend: mean max_abs_residual per month
    severity_per_month=event_summary.groupby("month")["max_abs_residual_MW"].mean().rename("avg_max_abs_residual_MW")

    ####################################################
    # Early detection metric vs z-score baseline
    # Baseline: rolling z-score of residual, same as earlier work
    ####################################################
    window=24*7  # 1 week
    res_mean = residual.rolling(window).mean()
    res_std  = residual.rolling(window).std()
    z = ((residual - res_mean) / res_std).rename("residual_z")
    z_flag = (np.abs(z) > 3).astype(int).rename("is_anomaly_zscore")
    # Build z-score events in the same way (for comparison)
    z_df = pd.DataFrame({"z_flag": z_flag}).dropna()
    z_times = z_df.index[z_df["z_flag"] == 1]
    if len(z_times) > 0:
        z_diffs = z_times.to_series().diff().dt.total_seconds().div(3600)
        z_new_group = (z_diffs.isna()) | (z_diffs > gap_tolerance_hours)
        z_group_id=z_new_group.cumsum().values

        z_only=pd.DataFrame(index=z_times)
        z_only["event_id"]=z_group_id
        z_events=(z_only.groupby("event_id").agg(start_time=("event_id", lambda s: s.index.min()),
                    end_time=("event_id", lambda s: s.index.max())).reset_index(drop=True))

        # Early detection: for each z-event, did Iso Forest flag earlier?? seems like
        # compute the first IF anomaly timestamp inside that event window in a small margin
        lead_times=[]
        for _, row in z_events.iterrows():
            start, end=row["start_time"], row["end_time"]
            # IF anomalies in the same window
            #if_window = anom_times[(anom_times >= start)]
            if_window=anom_times[(anom_times >= start)&(anom_times <= end)]
            lookback=pd.Timedelta(hours=24)
            if_window=anom_times[(anom_times >= start-lookback)&(anom_times <= end)]
            if len(if_window) > 0:
                lead_hours=(start-if_window.min()).total_seconds()/3600.0 #it is in seconds
                lead_hours=max(0.0, lead_hours)
                lead_times.append(lead_hours)      
        # Summarize early detection
        if len(lead_times) > 0:
            early_detection_mean_hours=float(np.mean(lead_times))
        else:
            early_detection_mean_hours=np.nan
    #Important to have this else here......
    else:
        early_detection_mean_hours=np.nan

    ####################################################
    #Useful plots
    ####################################################
    #residual with IF anomaly points
    plt.figure(figsize=(14,4))
    plt.plot(events_df.index, events_df["residual"], lw=0.8, label="Residual (Actual - Forecast)")
    #plt.scatter(anom_times, events_df.loc[anom_times, "residual"], color='red', label="IF anomalies")
    plt.scatter(anom_times, events_df.loc[anom_times, "residual"], color='red', s=12, label="IF anomalies")
    plt.title("Residuals with iso forests anomaly Points")
    plt.ylabel("Residuals(MW)")
    plt.legend()
    plt.show()
    # Score timeline (lower = more anomalous)
    plt.figure(figsize=(14,3))
    plt.plot(scores.index, scores.values, lw=0.8)
    plt.scatter(anom_times, scores.loc[anom_times], color='red', s=12)
    plt.title("Isolation forest score(lower = more anomalous)")
    plt.ylabel("IF score")
    plt.show()

    # Failure events per month
    plt.figure(figsize=(12,3))
    events_per_month.plot(kind="bar")
    plt.title("Forecast failure Events per Month (IsoForest)")
    plt.ylabel("Number of events")
    plt.tight_layout()
    plt.show()
    #Severity over time
    plt.figure(figsize=(12,3))
    severity_per_month.plot()
    plt.title("Average Event Severity (mean max abs(residual) per month)")
    plt.ylabel("Avg max abs(residual) (MW)")
    plt.tight_layout()
    plt.show()

    ####################################################
    #Metric summary
    ####################################################
    print("\n--- Metrics (Isolation Forest) ----")
    print(f"Total points evaluated: {len(events_df)}")
    print(f"Anomaly points flagged: {events_df['is_anomaly_iforest'].sum()} ({events_df['is_anomaly_iforest'].mean()*100:.2f}%)")
    print(f"Total anomaly events:  {len(event_summary)}")
    print("\nEvent duration (hours) summary:")
    print(duration_stats)
    print("\nTop 10 most severe events (by max abs(residual) in MW):")
    display(event_summary.sort_values("max_abs_residual_MW", ascending=False).head(10))
    print("\nEarly detection vs rolling z-score baseline:")
    print(f"Mean lead time (hours): {early_detection_mean_hours}")
```

```python
# DL approach, Pytorch Autoencoder
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
################################
## Train, test split
###############################
split = int(len(X)*0.80)
X_train_df=X.iloc[:split].copy()
X_test_df =X.iloc[split:].copy()
# At the moment, standardizing using train stats only..
mu=X_train_df.mean()
sd=X_train_df.std().replace(0, 1.0)
#X_train=((X_train_df-mu)/sd)
#X_test=((X_test_df-mu)/sd)
X_train=((X_train_df-mu)/sd).values.astype(np.float32)
X_test=((X_test_df-mu)/sd).values.astype(np.float32)

train_loader = DataLoader(TensorDataset(torch.from_numpy(X_train)),batch_size=200,shuffle=True)

################################
# Define encoder class
###############################
class AE(nn.Module):
    def __init__(self, d):
        #super().init()
        super().__init__()
        self.net = nn.Sequential(nn.Linear(d, 8),nn.ReLU(),nn.Linear(8, 2),nn.ReLU(), nn.Linear(2, 8),
            nn.ReLU(), # bottleneck=2 foces strong compression.. have tried running with increased bottleneck, it increases traced anamolous points
            nn.Linear(8, d),)
    #fwd pass    
    def forward(self, x):
        return self.net(x)
#device="cpu"
device="cuda" if torch.cuda.is_available() else "cpu"
model=AE(d=X_train.shape[1]).to(device)
opt=torch.optim.Adam(model.parameters(),lr=1e-3)
#loss function
loss_fn = nn.MSELoss()

#####Model training#####
model.train()
for epoch in range(8): #do a dedicated check to decide on the number of epochs..
    total = 0.0
    for (xb,) in train_loader:
        xb = xb.to(device)
        #recon=model(train_loader)
        recon = model(xb)
        loss = loss_fn(recon, xb)
        #opt.step()
        opt.zero_grad()
        loss.backward()
        opt.step()
        total=total+(loss.item()*len(xb))
    print(f"epoch {epoch+1}:train_mse={total/len(X_train):.3f}")

###score anamolies by rec error on TEST####
model.eval()
with torch.no_grad():
    #xt=torch.from_numpy(X_test)
    xt=torch.from_numpy(X_test).to(device)
    recon=model(xt).cpu().numpy()
recon_err=np.mean((recon-X_test)**2, axis=1)
scores=pd.Series(recon_err,index=X_test_df.index,name="AE_recon_mse")
###choose threshold using train error quantiles
with torch.no_grad():
    tr=torch.from_numpy(X_train).to(device)
    tr_recon=model(tr).cpu().numpy()
train_err=np.mean((tr_recon-X_train)** 2, axis=1)

contamination = 0.01  #match Isolation Forest setting
#Need to reconsider this threshold!!
#thr = float(np.quantile(train_err, 1-contamination)) #threshold on test..
thr = float(np.quantile(scores.values, 1-contamination)) # for fair comparison with iso forests
is_anom=(scores > thr).astype(int).rename("is_anomaly_ae")

##point anamolies mapped to events###
events_df=pd.DataFrame({"residual": residual.reindex(scores.index),"ae_score": scores,"is_anomaly_ae": is_anom})

gap_tolerance_hours=2
anom_times=events_df.index[events_df["is_anomaly_ae"] == 1]

print("Train anomaly rate at threshold:", (train_err > thr).mean())
print("Test  anomaly rate at threshold:", (scores.values > thr).mean())

if len(anom_times)==0:
    print("No anomalies detected. Try higher contamination!")
else:
    diffs=anom_times.to_series().diff().dt.total_seconds().div(3600)
    new_group=diffs.isna() | (diffs > gap_tolerance_hours)
    group_id=new_group.cumsum().values
    anom_only=events_df.loc[anom_times].copy()
    anom_only["event_id"]=group_id
    event_summary=(anom_only.groupby("event_id").agg(
        start_time=("residual", lambda s: s.index.min()),
        end_time=("residual", lambda s: s.index.max()),
        duration_hours=("residual", "count"),
        max_abs_residual_MW=("residual", lambda s: float(np.max(np.abs(s.values)))),
        mean_abs_residual_MW=("residual", lambda s: float(np.mean(np.abs(s.values)))),
        #min_score=("ae_score", "min"),
        max_score=("ae_score", "max"),
    ).sort_values("start_time").reset_index(drop=True))

    display(event_summary.head())

    event_summary["month"]=event_summary["start_time"].dt.to_period("M").dt.to_timestamp()
    #frequency: events per month
    events_per_month = event_summary.groupby("month").size().rename("events_per_month")
    #Duration: distribution + mean per month
    duration_stats=event_summary["duration_hours"].describe()
    #Severity trend: mean max_abs_residual per month
    severity_per_month= (event_summary.groupby("month")["max_abs_residual_MW"].mean().rename("avg_max_abs_residual_MW"))

    window=24*7  # 1 week
    res_mean=residual.rolling(window).mean()
    res_std=residual.rolling(window).std()
    z=((residual-res_mean)/res_std).rename("residual_z")
    z_flag = (np.abs(z) > 3).astype(int).rename("is_anomaly_zscore")
    #Build z-score events in the same way
    z_df=pd.DataFrame({"z_flag": z_flag}).dropna()
    z_times=z_df.index[z_df["z_flag"] == 1]

    if len(z_times) > 0:
        z_diffs=z_times.to_series().diff().dt.total_seconds().div(3600)
        z_new_group=(z_diffs.isna()) | (z_diffs > gap_tolerance_hours)
        z_group_id=z_new_group.cumsum().values
        z_only=pd.DataFrame(index=z_times)
        z_only["event_id"]=z_group_id
        z_events = (z_only.groupby("event_id").agg(
            start_time=("event_id", lambda s: s.index.min()),
            end_time=("event_id", lambda s: s.index.max()),)
        .reset_index(drop=True))
        #Early detection check for AE, same method as done for Iso Forests
        #look for AE anomaly points up to 24h before z-event start
        lookback = pd.Timedelta(hours=24)
        lead_times = []
        for _, row in z_events.iterrows():
            start, end=row["start_time"], row["end_time"]
            ae_window=anom_times[(anom_times >= start - lookback) & (anom_times <= end)]
            if len(ae_window) > 0:
               lead_hours=(start-ae_window.min()).total_seconds() / 3600.0
               lead_hours=max(0.0, lead_hours)
               lead_times.append(lead_hours)
        if len(lead_times) > 0:
          early_detection_mean_hours = float(np.mean(lead_times))
        else:
          early_detection_mean_hours = np.nan
    else:
        early_detection_mean_hours = np.nan

    #residual with AE anomaly points
    plt.figure(figsize=(14,4))
    plt.plot(events_df.index, events_df["residual"], lw=0.8, label="Residual(Actual-Forecast)")
    plt.scatter(anom_times, events_df.loc[anom_times, "residual"], color="red", s=12, label="AE anomalies")
    plt.title("residuals with anomaly points")
    plt.ylabel("residual (MW)")
    plt.legend()
    plt.show()
    #AE score timeline (higher = more anomalous, since recon error)
    plt.figure(figsize=(14,3))
    plt.plot(events_df.index, events_df["ae_score"], lw=0.8)
    plt.scatter(anom_times, events_df.loc[anom_times, "ae_score"], color="red", s=12)
    plt.title("reconstruction error (higher = more anomalous)")
    plt.ylabel("AE recon MSE")
    plt.show()

    #Failure events per month
    plt.figure(figsize=(12,3))
    events_per_month.plot(kind="bar")
    plt.title("Forecast failure events per month")
    plt.ylabel("Number of events")
    plt.tight_layout()
    plt.show()
    #Severity over time
    #plt.figure(figsize=(12,3))
    #severity_per_month.plot()
    #plt.title("Average event severity (mean max abs(residual) per month)")
    #plt.ylabel("Avg max abs(residual) (MW)")
    #plt.tight_layout()
    #plt.show()

   #############################
   #Metric summary AE
   ##############################
    print("\n---Metrics----")
    print(f"Total points evaluated (Test dataset): {len(events_df)}")
    print(f"Anomaly points flagged (Test dataset): {events_df['is_anomaly_ae'].sum()} ({events_df['is_anomaly_ae'].mean()*100:.2f}%)")
    print(f"Total anomaly events (Test dataset): {len(event_summary)}")
    print("\nEvent duration (hours) summary:")
    print(duration_stats)
    print("\nTop 10 most severe events in Test dataset (by max abs(residual) in MW):")
    display(event_summary.sort_values("max_abs_residual_MW", ascending=False).head(10))
    #print("\nEarly detection vs rolling z-score baseline:")
    #print(f"Mean lead time (hours): {early_detection_mean_hours}")
```
